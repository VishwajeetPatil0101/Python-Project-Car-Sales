Exploratory data analysis (EDA)

Total sales for all cars

```python
Total_sales = car_sales["Sales"].sum()
Total_sales
```

Top car sales By Each Manufactures

```python
Total_sales_each_manu = car_sales["Sales"].groupby(car_sales["Manufacturer"]).sum()
```

```python
fig, ax=plt.subplots()
ax.bar(Total_sales_each_manu.index,Total_sales_each_manu, color = "#E330F0")
ax.set_xlabel("Car Companies")
ax.set_ylabel("sales")
ax.set_title("Sales by Each Manufacturers")
plt.xticks(rotation=90)
plt.show;
```

```python
car_sales.describe()
```

Sales by Vehicle Type

```python
Total_sales_Type = car_sales["Sales"].groupby(car_sales["Vehicle_type"]).sum()
```

```python
Total_sales_Type
```

```python
fig, ax=plt.subplots()
ax.bar(Total_sales_Type.index,Total_sales_Type, color = "#54D1D1")
ax.set_xlabel("Vehicle Type")
ax.set_ylabel("sales")
ax.set_title("Sales by Vehicle Type")
plt.show;
```

Relationship Between Engine Size and Horsepower

```python
fig, ax=plt.subplots()
ax.scatter(car_sales["Engine_size"] , car_sales["Horsepower"], color = "#BF4226")
ax.set_xlabel("Engine Size")
ax.set_ylabel("Horsepower")
ax.set_title("Relationship Between Engine Size and Horsepower")
plt.show;
```

Relationship Between Wheelsize and Weight

```python
fig, ax=plt.subplots()
ax.scatter(car_sales["Wheelbase"] , car_sales["Curb_weight"], color = "#BF4226")
ax.set_xlabel("Wheelbase ")
ax.set_ylabel("weight")
ax.set_title("Relationship Between Wheelsize and Weight")
plt.show;
```

```python
fig, ax=plt.subplots()
ax.scatter(car_sales["Fuel_efficiency"] , car_sales["Power_perf_factor"], color = "#DB0272")
ax.set_xlabel("Fuel efficiency ")
ax.set_ylabel("Power_perf_factor")
ax.set_title("Relationship Between Fuel Efficiency and Power Factor")
plt.show;
```
